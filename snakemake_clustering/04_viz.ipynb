```r
library(ggplot2)
library(cowplot)
library(dplyr)
library(cowplot)
```

```r
df <- read.csv('../../results/05_snakemake_clustering/03_eva_clustering/ari.csv', row.names = 1)
```

```r
head(df)
```

```r
#df <- subset(df, clustering_method != "Kmeans")
```

```r
options(repr.plot.height = 6, repr.plot.width = 8)

p <- df %>% subset(n_gene == 500) %>%
    ggplot(aes(x = reorder(svg_method, -ari, median), y = ari)) +
    geom_boxplot() +
    ggtitle("500 genes") +
    theme_cowplot() +
    theme(axis.text.x = element_text(angle = 60, hjust = 1))

print(p)
```
